```python
import pandas as pd

data = pd.read_csv('starbucks_drinkMenu_expanded.csv', header=0)
print(data.head())
```

```python
data.columns
```

```python
data.rename(columns={
       ' Total Fat (g)': 'Fat', 'Trans Fat (g) ':'TransFat', 'Saturated Fat (g)':'SatFat', ' Sodium (mg)':'mgSodium',
       ' Total Carbohydrates (g) ':'Carbs', 'Cholesterol (mg)':'mgCholesterol', ' Dietary Fibre (g)':'Fiber',
       ' Sugars (g)':'Sugar',' Protein (g) ':'Protein', 'Vitamin A (% DV) ':'VitA', 'Vitamin C (% DV)':'VitC',
       ' Calcium (% DV) ':'Calcium', 'Iron (% DV) ':'Iron', 'Caffeine (mg)':'mgCaffeine'}, errors='raise', inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['VitA'] = list(map(lambda x: x[:-1], data['VitA'].values))
```

```python
data['VitC'] = list(map(lambda x: x[:-1], data['VitC'].values))
data['Calcium'] = list(map(lambda x: x[:-1], data['Calcium'].values))
data['Iron'] = list(map(lambda x: x[:-1], data['Iron'].values))
```

```python
data['VitA'] = [float(x) for x in data['VitA'].values]
data['VitC'] = [float(x) for x in data['VitC'].values]
data['Calcium'] = [float(x) for x in data['Calcium'].values]
data['Iron'] = [float(x) for x in data['Iron'].values]
```

```python
data['Fat'].unique()
```

```python
data['Fat'].replace('3 2', '3', inplace=True)
```

```python
data['Fat'] = [float(x) for x in data['Fat'].values]
```

```python
data['mgCaffeine'].unique()
```

```python
data['mgCaffeine'].replace('90%', '90', inplace=True)
```

```python
data['mgCaffeine'] = [float(x) for x in data['mgCaffeine'].values]
```

```python
data.info()
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder 
from sklearn.cluster import KMeans
```

```python
##Calories v fat
sns.scatterplot(data=data, x="Fat", y="Calories", hue="Milk", size="Size", size_order=['Venti', 'Grande', 'Tall', 'Short'],sizes = (50, 150))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Calories vs. Fat in Starbucks Drinks')
```

```python
corr = data.corr()
sns.heatmap(corr)
```

```python
sns.pairplot(data)
```

```python
##Bar chart of calories
ax = sns.barplot(x="Beverage_category", y="Calories", data=data)
ax.set_xticklabels(
    labels=["Coffee", "Espresso Drinks", "Signature Espresso Drinks", "Tea", 'Shaken Iced', 'Smoothies', 'Frapp', 'Frapp Light', 'Frapp Creme'], rotation=80)
plt.title('Calories by beverage type at Starbucks')
```

```python
g = sns.FacetGrid(col='Beverage_category',data=data,legend_out=False)
g.map(sns.boxplot,'Calories')
```

```python
sns.scatterplot(data=data, x="Calories", y="mgCaffeine", hue="Milk", size="Size", size_order=['Venti', 'Grande', 'Tall', 'Short'],sizes = (50, 150))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Caffeine vs. Calories in Starbucks Drinks')
plt.xlabel('mg of Caffeine per drink')
plt.ylabel('Calories per drink')
```
